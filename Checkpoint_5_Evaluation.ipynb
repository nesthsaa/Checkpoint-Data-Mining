###**Penanganan Outliers**

#####**Pengecekkan Outliers**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

#####**Hapus Outliers**

Melakukan penghapusan baris pada kolom "reviewsCount" karena memiliki persentase <= 10% untuk menanganani outliers

```python
Q1 = df['reviewsCount'].quantile(0.25)
Q3 = df['reviewsCount'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['reviewsCount'] >= lower_bound) & (df['reviewsCount'] <= upper_bound)]
```

# **Evaluation**

## **Evaluate Result (Unsupervised Learning)**

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

silhouette_avg_kmeans = silhouette_score(x_final_norm, kmeans.labels_)
ch_score_kmeans = calinski_harabasz_score(x_final_norm, kmeans.labels_)
db_score_kmeans = davies_bouldin_score(x_final_norm, kmeans.labels_)

silhouette_avg_hc = silhouette_score(x_final_norm, hierarchy_clust)
ch_score_hc = calinski_harabasz_score(x_final_norm, hierarchy_clust)
db_score_hc = davies_bouldin_score(x_final_norm, hierarchy_clust)

metrics = ["Silhouette Score", "Calinski-Harabasz Score", "Davies-Bouldin Score"]
scores_kmeans = [silhouette_avg_kmeans, ch_score_kmeans, db_score_kmeans]
scores_hc = [silhouette_avg_hc, ch_score_hc, db_score_hc]

df_scores = pd.DataFrame({'': metrics,
                          'KMeans Score': scores_kmeans,
                          'Agglomerative Clustering Score': scores_hc})

df_scores = df_scores.set_index('')

df_scores = df_scores.T

display(df_scores)
```

Dalam bagian ini, dilakukan evaluasi hasil dari proses model unsupervised learning. Digunakan beberapa metrik evaluasi untuk mengevaluasi seberapa baik model klastering (KMeans dan Agglomerative Clustering) dalam mengelompokkan data.

Metrik evaluasi yang digunakan meliputi:
- **Silhouette Score**: Metrik ini mengukur seberapa baik setiap sampel ditempatkan dalam klasternya sendiri dibandingkan dengan klaster lainnya. Nilai yang lebih tinggi menunjukkan bahwa klaster memiliki pemisahan yang lebih baik.
- **Calinski-Harabasz Score**: Juga dikenal sebagai indeks kekompakan, metrik ini mengukur seberapa kompak klaster-klasternya dan seberapa terpisah satu sama lain. Nilai yang lebih tinggi menunjukkan klaster yang lebih baik.
- **Davies-Bouldin Score**: Metrik ini mengukur seberapa baik klaster-klasternya didefinisikan. Nilai yang lebih rendah menunjukkan klaster yang lebih baik.

Skor evaluasi dihitung berdasarkan kedua model klastering (KMeans dan Agglomerative Clustering) yang telah dilatih sebelumnya pada data `x_final_norm`.

#### Membuat Model KMeans dengan cluster n = 8

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Inisialisasi dan melatih model KMeans
kmeans = KMeans(n_clusters=8, random_state=42)
kmeans.fit(x_final_norm)  # Saya mengganti X_train dengan x_final_norm karena saya tidak melihat deklarasi X_train sebelumnya

# Dapatkan label klaster yang diprediksi oleh model
cluster_labels = kmeans.labels_

# Reduksi dimensi data dengan PCA
pca = PCA(n_components=2)
x_final_pca = pca.fit_transform(x_final_norm)

# Gabungkan label klaster dengan komponen utama
cluster_df = pd.DataFrame(x_final_pca, columns=['PC1', 'PC2'])
cluster_df['Cluster'] = cluster_labels

# Visualisasi
plt.figure(figsize=(10, 6))
for cluster in range(8):
    plt.scatter(cluster_df.loc[cluster_df['Cluster'] == cluster, 'PC1'],
                cluster_df.loc[cluster_df['Cluster'] == cluster, 'PC2'],
                label=f'Cluster {cluster}', alpha=0.5)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Visualization of Clusters')
plt.legend()
plt.show()
```

Membagi model KMeans dengan 8 klaster dan dipelajari (dilatih) menggunakan data x_final_norm. Ini berarti model akan mencoba mengelompokkan data menjadi 8 klaster berdasarkan pola yang ada dalam data tersebut.

#### Membuat Model Agglomerative dengan cluster n = 3

## **Review Process**

### Cross Validation

```python
from sklearn.model_selection import KFold
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

n_splits = 5
kf = KFold(n_splits=n_splits)

kmeans_scores = {'Silhouette Score': [], 'Calinski-Harabasz Score': [], 'Davies-Bouldin Score': []}
hc_scores = {'Silhouette Score': [], 'Calinski-Harabasz Score': [], 'Davies-Bouldin Score': []}

# KMeans
for train_index, test_index in kf.split(x_final_norm):
    X_train_fold = x_final_norm[train_index]

    kmeans.fit(X_train_fold)
    kmeans_labels = kmeans.labels_

    silhouette_avg = silhouette_score(X_train_fold, kmeans_labels)
    ch_score = calinski_harabasz_score(X_train_fold, kmeans_labels)
    db_score = davies_bouldin_score(X_train_fold, kmeans_labels)

    kmeans_scores['Silhouette Score'].append(silhouette_avg)
    kmeans_scores['Calinski-Harabasz Score'].append(ch_score)
    kmeans_scores['Davies-Bouldin Score'].append(db_score)

# Agglomerative Clustering
for train_index, test_index in kf.split(x_final_norm):
    X_train_fold = x_final_norm[train_index]

    hc.fit(X_train_fold)
    hc_clusters = hc.labels_

    silhouette_avg = silhouette_score(X_train_fold, hc_clusters)
    ch_score = calinski_harabasz_score(X_train_fold, hc_clusters)
    db_score = davies_bouldin_score(X_train_fold, hc_clusters)

    hc_scores['Silhouette Score'].append(silhouette_avg)
    hc_scores['Calinski-Harabasz Score'].append(ch_score)
    hc_scores['Davies-Bouldin Score'].append(db_score)

# Hitung rata-rata skor
avg_kmeans_scores = {metric: sum(values) / len(values) for metric, values in kmeans_scores.items()}
avg_hc_scores = {metric: sum(values) / len(values) for metric, values in hc_scores.items()}

print("Average KMeans Scores:")
print(avg_kmeans_scores)
print("\nAverage Agglomerative Clustering Scores:")
print(avg_hc_scores)
```

1. **Silhouette Score**:
   - KMeans memiliki nilai Silhouette Score sekitar 0.083, sedangkan Agglomerative Clustering memiliki nilai sekitar 0.066.
   - Nilai Silhouette Score KMeans lebih tinggi, menunjukkan bahwa KMeans memberikan pemisahan klaster yang sedikit lebih baik dibandingkan Agglomerative Clustering.

2. **Calinski-Harabasz Score**:
   - KMeans memiliki nilai Calinski-Harabasz Score sekitar 11.03, sedangkan Agglomerative Clustering memiliki nilai sekitar 12.99.
   - Nilai Calinski-Harabasz Score Agglomerative Clustering lebih tinggi, menunjukkan bahwa Agglomerative Clustering memberikan klaster yang lebih kompak dan terpisah dengan baik.

3. **Davies-Bouldin Score**:
   - KMeans memiliki nilai Davies-Bouldin Score sekitar 3.05, sedangkan Agglomerative Clustering memiliki nilai sekitar 1.27.
   - Nilai Davies-Bouldin Score Agglomerative Clustering lebih rendah, menunjukkan bahwa Agglomerative Clustering  memberikan pemisahan klaster yang lebih baik.
